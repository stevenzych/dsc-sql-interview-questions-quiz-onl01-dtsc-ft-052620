## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

#### One-to-one, one-to-many, and many-to-many. Let's say we are tracking recipes at our restaurant in a database. An example of one-to-one would be a recipe that calls for only one ingredient, and that one ingredient is ONLY used in that dish (sounds like a bad recipe). One-to-many could be one ingredient that can be put in many different recipes. And a many-to-many relationship could be a bunch of recipes that all call for different combinations of many ingredients, some of which are overlapping.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
#### SQL has four main joins: Left, where everything from the left table and matching records from the right are kept; Right, where the inverse is true; Inner, where only records that exist in both tables are kept; and Full, where all records from both tables are kept.
